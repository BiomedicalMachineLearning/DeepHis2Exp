```python
import matplotlib.pyplot as plt
import os
import gc
```

```python
# List of figure names
figure_names = ['BC_Her2ST_ID.png', 'Kidney_visium_ID.png', 'Skin_cSCC_ID.png',
                'BC_visium_ID.png', 'Liver_visium_ID.png', 'Skin_Melanoma_ID.png']
title_names = ['BC_Her2ST.png', 'Kidney_visium.png', 'Skin_cSCC.png',
                'BC_visium.png', 'Liver_visium.png', 'Skin_Melanoma.png']
# Calculate the figure size for A4 paper (in inches)
a4_width = 8.27
a4_height = 11.69
fig_width = 3 * a4_width  # 3x2 grid
fig_height = 2 * a4_height

# Create a 3x2 grid of subplots
fig, axs = plt.subplots(3, 2, figsize=(fig_width, fig_height))
fig.suptitle('Gene enrichment analysis', fontsize=32)  # Main title for all subplots

# Iterate over each subplot and plot the corresponding figure
for i, ax in enumerate(axs.flatten()):
    if i < len(figure_names):
        figure_path = os.path.join('./', figure_names[i])
        img = plt.imread(figure_path)
        ax.imshow(img)
        ax.axis('off')  # Turn off axis labels
        ax.set_title(os.path.splitext(title_names[i])[0], fontsize=24)  # Use filename as subtitle with adjusted font size

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout and add a bit of space for the main title
plt.savefig("S2.pdf")
plt.show()
gc.collect()
```
