# Load latent space

```python
import pickle
# Read dictionary pkl file
with open('swim_t.pkl', 'rb') as fp:
    features = pickle.load(fp)
with open('label(33samples).pkl', 'rb') as fp:
    labels = pickle.load(fp)    
    
data, target = features, labels
```

```python
latent_space = data["A1"].numpy()
latent_space.shape
```

# PCA Plot

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Perform PCA to reduce the dimensionality of the latent space representation to 2 dimensions
pca = PCA(n_components=7)
latent_space_pca = pca.fit_transform(latent_space)

# Plot the reduced latent space representation using PCA
plt.figure(figsize=(5, 5))
plt.scatter(latent_space_pca[:, 0], latent_space_pca[:, 1])
plt.title('PCA Visualization of Latent Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import numpy as np
# Obtain the explained variance ratio for each component
explained_var = pca.explained_variance_ratio_

# Compute the cumulative explained variance
cumulative_var = np.cumsum(explained_var)

# Plot the cumulative explained variance as a function of the number of components
plt.plot(cumulative_var)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

# T-SNE Plot

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Perform t-SNE to reduce the dimensionality of the latent space representation to 2 dimensions
tsne = TSNE(n_components=2)
latent_space_tsne = tsne.fit_transform(latent_space)

# Plot the reduced latent space representation using t-SNE
plt.figure(figsize=(5, 5))
plt.scatter(latent_space_tsne[:, 0], latent_space_tsne[:, 1])
plt.title('t-SNE Visualization of Latent Space')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```

# UMAP Plot

```python
import umap.umap_ as umap

# Perform UMAP to reduce the dimensionality of the latent space representation to 2 dimensions
umap_model = umap.UMAP(n_components=7)
latent_space_umap = umap_model.fit_transform(latent_space)

import matplotlib.pyplot as plt

# Plot the reduced latent space representation using UMAP
plt.figure(figsize=(5, 5))
plt.scatter(latent_space_umap[:, 0], latent_space_umap[:, 1])
plt.title('UMAP Visualization of Latent Space')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.show()
```
