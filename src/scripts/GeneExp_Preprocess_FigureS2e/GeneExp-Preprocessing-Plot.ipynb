```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("GeneExp-Preprocessing.csv")

plt.figure(figsize=(9, 6))
ax=sns.boxplot(data=df, x="Dataset", y="Pearson correlation", hue="Preprocessing Methods", showfliers=False)
plt.ylabel("PCC")
sns.set(font_scale=1.6)
sns.set_style("white")
plt.xticks([0, 1], ['HER2+ Dataset', 'Visium Dataset'])

sns.move_legend(ax, "best", bbox_to_anchor=(1, 1))
ax.set(xlabel=None)
```
